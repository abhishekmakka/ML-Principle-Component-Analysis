```python
#Importing Libraries 
import numpy as np
import pandas as pd

#importing visualization
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#importing data
data = pd.read_csv('Wine.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
#checking for null values
data.isnull().sum()
```

```python
#importing scale and normalize
from sklearn.preprocessing import scale, normalize
```

```python
data_scale = scale(data)
```

```python
data_scale
```

```python
data_normalize = normalize(data_scale)
```

```python
data_normalize
```

```python
#importing principal component analysis
from sklearn.decomposition import PCA
```

```python
model = PCA()
```

```python
model.fit(data_normalize)
```

```python
model.components_
```

```python
#converting 1D to 2D
Componets = pd.DataFrame(model.components_)
```

```python
Componets 
```

```python
value = model.explained_variance_ratio_
```

```python
value.sum()
```

```python
value.mean()
```
